```python
%matplotlib inline
import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt

room = cv2.imread('room1.jpg')
# print(room_img)

room = cv2.cvtColor(room, cv2.COLOR_BGR2RGB)

# plt.imshow(room_img)
# plt.show()

room_gray = cv2.cvtColor(room, cv2.COLOR_RGB2GRAY)

# print(room_gray)

# plt.imshow(room_gray)
# plt.show()

room_preprocessed  = cv2.GaussianBlur(room_gray, (5, 5), 0)
# type(room_preprocessed)

# get canny binary
_, room_thresh = cv2.threshold(room_preprocessed, 80, 255, cv2.THRESH_BINARY)

plt.imshow(cv2.cvtColor(room_thresh, cv2.COLOR_GRAY2RGB))

room_edges = cv2.Canny(room_preprocessed, threshold1=90, threshold2=110)
plt.imshow(cv2.cvtColor(room_edges, cv2.COLOR_GRAY2RGB))
cv2.imwrite('room-edges.jpg', room_edges)
```
