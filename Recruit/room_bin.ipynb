```python
%matplotlib inline
import cv2, matplotlib
import numpy as np
import matplotlib.pyplot as plt

room = cv2.imread('room1.jpg')
# print(room_img)

room = cv2.cvtColor(room, cv2.COLOR_BGR2RGB)

# plt.imshow(room_img)
# plt.show()

room_gray = cv2.cvtColor(room, cv2.COLOR_RGB2GRAY)

# print(room_gray)

# plt.imshow(room_gray)
# plt.show()

room_preprocessed  = cv2.GaussianBlur(room_gray, (5, 5), 0)
# type(room_preprocessed)

# bet binary image
# threshold 10-255 -> 黒枠だけ取られる
# threshold 170-255 -> 細い枠，文字まで
_, room_binary = cv2.threshold(room_preprocessed, 10, 255, cv2.THRESH_BINARY)
room_binary = cv2.bitwise_not(room_binary)

plt.imshow(cv2.cvtColor(room_binary, cv2.COLOR_GRAY2RGB))
cv2.imwrite('room-binary.jpg', room_binary)

# copy of image to draw lines
room_line = np.copy(room)

# find hough lines
num_pix_threshold = 100
lines = cv2.HoughLines(room_binary, 1, np.pi/180, num_pix_threshold)

```
